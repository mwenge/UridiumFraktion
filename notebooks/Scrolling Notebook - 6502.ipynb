### Use pixelX and segment to calculate address position

```z80
scrollPositionLoPtr = $10
scrollPositionHiPtr = $11
maskForHiPtr = $0F
surfaceDataForCurrentLevel = $8200
pixelX = $09
segment = $14

LDA #214
STA pixelX
LDA #03
STA segment

LDA pixelX
CLC
ADC #$07
STA scrollPositionLoPtr

LDA segment
ADC #$00
LSR
ROR scrollPositionLoPtr
LSR
ROR scrollPositionLoPtr
LSR
ROR scrollPositionLoPtr
AND #$01
STA maskForHiPtr

LDA #>surfaceDataForCurrentLevel
ORA maskForHiPtr
STA scrollPositionHiPtr
```
